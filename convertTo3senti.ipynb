```python
import pandas as pd
```

```python
data = pd.read_csv("Dataset/sentiment_data_preprocessed.csv")
data
```

```python
def sentiment(x):
    if x>0.6:
        return "positive"
    if x<0.4:
        return "negative"
    return "neutral"
```

```python
data['implication'] = data['sentiment_score'].apply(sentiment)
```
